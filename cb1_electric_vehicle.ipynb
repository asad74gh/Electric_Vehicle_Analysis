 List down the top 10 states that had the highest compounded annual

```python
df_2022s
```

```python
state22=state22[1:].set_index('state')
```

```python
state22
```

```python
state24=df_2024s.groupby(['state'])['total_vehicles_sold'].sum().reset_index()
```

```python
state24=state24.rename(columns={'total_vehicles_sold':'tvs24'})
```

```python
state24=state24[1:].set_index('state')
```

```python
state24
```

```python
df22=pd.DataFrame(state22)
```

```python
df24=pd.DataFrame(state24)
```

```python
frame=[df22,df24]
```

```python
df11=pd.concat(frame,axis=1,join='inner')
```
